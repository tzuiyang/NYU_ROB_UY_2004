## Start the Visualizer

````python
from vis_tool import Visualizer
vis = Visualizer()
````

If above is successful, you should see the visualizer window pop up. Given this visualizer, you can show coordinate frames as you wish. But let's first add a coordinate frame at origin:

````python
import numpy as np
vis.logCoordinateFrame(np.eye(4), 'origin')
````

Now that you have a frame at origin, let's add a frame at (0, 0, 0.3). If you remember, the 4x4 transformation matrix representing this motion is:

````python
world_T_1 = np.array([
    [1,     0,      0,      0.3],
    [0,     1,      0,      0.0],
    [0,     0,      1,      0.0],
])
world_T_1
vis.logCoordinateFrame(world_T_1, 'origin/frame1')
````
